The practical implication of this value is that one-unit increase in 'AT' means an average 2.1713-unit decrease in 'EP'. It is clear from the regression scatterplot that my previous assertions is correct, model fits well, and AT' and 'PE' are negatively strong correlated.

From the plot we can see there are no outliers of this model need to remove from the data

### Improve model ---- using possible interaction terms or nonlinear assocations



```python
#split training data and testing data from original dataset
x=df.iloc[:,0:4]
y=df.iloc[:,[4]]
training=pd.concat([x,y],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=0)

#normal linear regression
x_train=sm.add_constant(x_train)
model=sm.OLS(y_train,x_train)
results=model.fit()
print(results.summary())
```

```python
#linear regression MSE
y_train_pred = results.predict(x_train)
train_error = mean_squared_error(y_train, y_train_pred)
print('train MSE:',train_error)

x_test=sm.add_constant(x_test)
y_test_pred = results.predict(x_test)
test_error = mean_squared_error(y_test, y_test_pred)
print('test MSE:',test_error)
```

```python
#involving all possible interaction terms and quadratic nonlinearities
x = df[['AT', 'V', 'AP', 'RH']]
quadratic = PolynomialFeatures(degree=2)
ndf = quadratic.fit_transform(x)
ndf = pd.DataFrame(ndf, columns = ['Coff','AT', 'V', 'AP', 'RH', 'AT2', 'AT*V', 'AT*AP', 'AT*RH', 'V2', 'V*AP', 'V*RH', 'AP2', 'AP*RH', 'RH2'])
ndf['PE'] = df['PE']

x=ndf.iloc[:,1:15]
y=ndf.iloc[:,[15]]
training=pd.concat([x,y],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=0)
```

```python
#build model
model = smf.ols("PE ~ 1 + AT + V + AP + RH + AT2 + AT*V + AT*AP + AT*RH + V2 + V*AP + V*RH + AP2 + AP*RH + RH2", training) 
results = model.fit()
print(results.summary())
```

```python
#MSE
x_train=sm.add_constant(x_train)
y_train_pred = results.predict(x_train)
train_error = mean_squared_error(y_train, y_train_pred)
print('train MSE:',train_error)

x_test=sm.add_constant(x_test)
y_test_pred = results.predict(x_test)
test_error = mean_squared_error(y_test, y_test_pred)
print('test MSE:',test_error)
```

```python
#remove insiginificant varibles
model = smf.ols("PE ~ 1 + AT + V + AP + RH + AT2 + AT*V + AT*RH + AP2 + AP*RH + RH2", training) 
results = model.fit()
print(results.summary())
```

```python
#new MSE
x_train=sm.add_constant(x_train)
y_train_pred = results.predict(x_train)
train_error = mean_squared_error(y_train, y_train_pred)
print('train MSE:',train_error)


x_test=sm.add_constant(x_test)
y_test_pred = results.predict(x_test)
test_error = mean_squared_error(y_test, y_test_pred)
print('test MSE:',test_error)
```

### KNN Regression

```python
#raw data
x=df.iloc[:,0:4]
y=df.iloc[:,[4]]
training=pd.concat([x,y],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=0)


#MSE and best k
training_error_array=[]
test_error_array=[]
k_list=[]
for k in range(1,101):
    k_list.append(k)
    knn = KNeighborsRegressor(n_neighbors=k)
    model=knn.fit(x_train,y_train)
    
    y_train_pred=knn.predict(x_train)
    y_test_pred=knn.predict(x_test)
    
    knn_train_error=mean_squared_error(y_train,y_train_pred)
    knn_test_error=mean_squared_error(y_test,y_test_pred)
    
    test_error_array.append(knn_test_error)
    training_error_array.append(knn_train_error)

min_test_error=min(test_error_array)
p=test_error_array.index(min_test_error)
best_k=k_list[p]
print('best k is',best_k)
print('test MSE:',min_test_error)


#plot test error and training error
k_list=[(1/k) for k in range(1,101)]
plt.plot(k_list,test_error_array,label='test_error_rate',color='green')
plt.plot(k_list,training_error_array,label='training_error_rate',color='red')
plt.xlabel('k')
plt.ylabel('error_rate')
plt.legend()
print('                 train and test error')
plt.show()
```

```python
#normalized data
norm_df = preprocessing.normalize(df)
norm_df = pd.DataFrame(data=norm_df, columns=['AT','V','AP','RH','PE'])
x=norm_df.iloc[:,0:4]
y=norm_df.iloc[:,[4]]
training=pd.concat([x,y],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=0)


#MSE and best k
training_error_array=[]
test_error_array=[]
k_list=[]
for k in range(1,101):
    k_list.append(k)
    knn = KNeighborsRegressor(n_neighbors=k)
    model=knn.fit(x_train,y_train)
    
    y_train_pred=knn.predict(x_train)
    y_test_pred=knn.predict(x_test)
    
    knn_train_error=mean_squared_error(y_train,y_train_pred)
    knn_test_error=mean_squared_error(y_test,y_test_pred)
    
    test_error_array.append(knn_test_error)
    training_error_array.append(knn_train_error)

min_test_error=min(test_error_array)
p=test_error_array.index(min_test_error)
best_k=k_list[p]
print('best k is',best_k)
print('test MSE:',min_test_error)


#plot test error and training error
k_list=[(1/k) for k in range(1,101)]
plt.plot(k_list,test_error_array,label='test_error_rate',color='green')
plt.plot(k_list,training_error_array,label='training_error_rate',color='red')
plt.xlabel('k')
plt.ylabel('error_rate')
plt.legend()
print('                   train and test error')
plt.show()
```

b: small number of predictor result in overfitting, so inflexible model is better

c: when the relationship between the predictors and response is highly non-linear, flexible model is better 

d: when 2 = Var() is large, there are many noise, so inflexible model is better
